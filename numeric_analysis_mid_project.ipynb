```python
def NR(f, x0, eps=pow(10, -15), step=5000):
    # The input is symbol function (string) so we can calculate it's derivative
    fderivative = f.diff(x)
    for n in range(step):
        if abs(f.evalf(subs= {x:x0})) < eps:
            print(f"newton raphson method converged to {x0} in {n} steps")
            return
        else:
            x0 = x0 - np.float(f.evalf(subs= {x:x0})) / np.float(fderivative.evalf(subs= {x:x0}))  
    print("newton raphson method didn't converge")
    return


def bisection(f, x1, x2, eps=pow(10, -15), step=5000):
    # The input f is a function
    if f(x1) * f(x2) > 0:
        print("Both points in f result negative")
        return
    else:
        if f(x1) < 0:
            a = x1
            b = x2
        else:
            b = x1
            a = x2
    c = (a + b) / 2
    for n in range(step):
        if abs(f(c)) < eps:
            print(f"bisection method converged to {c} in {n} steps")
            return
        if f(c) < 0:
            a = c
        else:
            b = c
        c = (a + b) / 2
    print("bisection method didn't converge")
    return
```

```python
def newf(n):
  mult = 1
  for i in range(n):
    mult *= (n / (n-i))
  return mult

for n in range (1, 1001):
  print(f"({n}, {newf(n)})")
```

```python
# part 2
def A(eps):
  return [[1, eps], [1, 1]]

# calculate norms for matrix A
eps = [(i/10) for i in range(10)]
y_1norm = [np.linalg.norm(x=A(i), ord=1) for i in eps]
y_2norm = [np.linalg.norm(x=A(i), ord=2) for i in eps]
y_frob = [np.linalg.norm(x=A(i), ord='fro') for i in eps]
y_inf = [np.linalg.norm(x=A(i), ord=np.inf) for i in eps]
y_2norm_sqr2 = [(np.sqrt(2) * np.linalg.norm(x=A(i), ord=2)) for i in eps]

fig, ax = plt.subplots()
ax.plot(eps, y_1norm, label='1 norm')
ax.plot(eps, y_2norm, label='2 norm')
ax.plot(eps, y_inf, label='inf norm')
ax.plot(eps, y_frob, label='Frobenius norm')
ax.plot(eps, y_2norm_sqr2, label='2 norm mult sqrt(2)')
plt.xlabel("Epsilon")
plt.ylabel("Norm")
legend = ax.legend(loc='upper left')
plt.show()
```

```python
# part 3
cond_1norm = [np.linalg.cond(x=A(i), p=1) for i in eps]
cond_2norm = [np.linalg.cond(x=A(i), p=2) for i in eps]
cond_frob = [np.linalg.cond(x=A(i), p='fro') for i in eps]
cond_inf = [np.linalg.cond(x=A(i), p=np.inf) for i in eps]

fig, ax = plt.subplots()
ax.plot(eps, cond_1norm, label='1 norm')
ax.plot(eps, cond_2norm, label='2 norm')
ax.plot(eps, cond_frob, label='Frobenius norm')
ax.plot(eps, cond_inf, label='inf norm')
plt.xlabel("Epsilon")
plt.ylabel("Cond")
legend = ax.legend(loc='upper left')
plt.show()
```

```python
print(f"When epsilon equals {eps[0]}")
print("Cond for Norm 1:", cond_1norm[0])
print("Cond for Norm 2:", cond_2norm[0])
print("Cond for Norm inf:", cond_inf[0])
print("Cond for Norm Frobenius:", cond_frob[0])
```

```python
#part 3, a
def A(alpha):
  return [[10, 20, 2], [1, 1 + alpha, 4], [0, 100, 1]]

def L_no_pivot(alpha):
  return [[1, 0, 0], [0.1, 1, 0], [0, 100 / (alpha - 1), 1]]

def U_no_pivot(alpha):
  return [[10, 20, 2], [0, alpha - 1, 19 / 5], [0, 0, 1 - (380 / (alpha - 1))]]

def L_with_pivot():
  return [[1, 0, 0], [0, 1, 0], [0.1, 0, 1]]

def U_with_pivot():
  return [[10, 20, 2], [0, 100, 1], [0, 0, 19 / 5]]

l = np.linalg.cond(x=L_with_pivot(), p=1)
u = np.linalg.cond(x=U_with_pivot(), p=1)
print(f"With norm 1 - L cond: {l}, U cond: {u}, and multiply them: {l*u}")

l = np.linalg.cond(x=L_with_pivot(), p=2)
u = np.linalg.cond(x=U_with_pivot(), p=2)
print(f"With norm 2 - L cond: {l}, U cond: {u}, and multiply them: {l*u}")
```

```python
# part 3, c
def new_L_with_pivot(alpha):
  return [[1, 0, 0], [0, 1, 0], [0.1, (alpha - 1) / 100, 1]]

def new_U_with_pivot(alpha):
  return [[10, 20, 2], [0, 100, 1], [0, 0, (19 / 5) - ((alpha - 1) / 100)]]

alpha = [0.1, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 4.9]
y_1bound = [np.linalg.cond(x=new_L_with_pivot(i), p=1) * np.linalg.cond(x=new_U_with_pivot(i), p=1) for i in alpha]
y_2bound = [np.linalg.cond(x=new_L_with_pivot(i), p=2) * np.linalg.cond(x=new_U_with_pivot(i), p=2) for i in alpha]
y_1cond = [np.linalg.cond(x=A(i), p=1) for i in alpha]
y_2cond = [np.linalg.cond(x=A(i), p=2) for i in alpha]

fig, ax = plt.subplots()
ax.plot(alpha, y_1cond, label='1 norm cond(A)')
ax.plot(alpha, y_2cond, label='2 norm cond(A)')
ax.plot(alpha, y_1bound, label='1 norm bound')
ax.plot(alpha, y_2bound, label='2 norm bound')
plt.xlabel("Alpha")
plt.ylabel("Cond")
legend = ax.legend(loc='upper right')
plt.show()
```
